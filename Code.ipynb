```python
#Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
#Reading the data 
data = pd.read_csv("athlete_events.csv")
```

```python
#Printing the data
data
```

```python
#Describing the dataset
data.describe()
```

```python
#Printing the dimensions of dataset
print("Dataset Dimensions\nRows : {0}\nColumns : {1}".format(data.shape[0],data.shape[1])) 
```

```python
#Cleaning the data frame
#Data Frame before dropping the columns
data
```

```python
#Dropping the passed columns from the data frame
data.drop(["ID","Games","City","Event"],axis=1,inplace=True)
#Displaying the Data frame
data
```

```python
#Finding all the null values
data.isnull().sum()
```

```python
#Visualization of null values in dataset using heat map
sb.heatmap(data.isnull(), cbar=False)
plt.show()
```

```python
# Imputing the null values with mean values of their respective columns

data["Medal"].fillna("No Medal", inplace = True)
data["Age"].fillna(data["Age"].mean(), inplace = True)
data["Height"].fillna(data["Height"].mean(), inplace = True)
data["Weight"].fillna(data["Weight"].mean(), inplace = True)


#for col in data.columns[data.isnull().any()]:
    #col = pd.Series(col)
    #data[col].fillna(data[col].mean(), inplace = True)
```

```python
# Sorting data for female participation in the Olympics.

womenInOlympics = data[data.Sex == 'F']

#Sorting data for male participation in the Olympics.

menInOlympics = data[data.Sex == 'M']

#Sorting data for female gold medallists above the age of 25 in the Olympics.

womenGoldAboveTwentyfive = data[(data.Sex == 'F') & (data.Medal == "Gold") & (data.Age > 25)]

# Sorting data for male gold medallists above the age of 30 in the Olympics.

menGoldAboveThirty = data[(data.Sex == 'M') & (data.Medal == "Gold") & (data.Age > 30)]

# Sorting data for medals won by India in the Olympics. 

medalsIndia = data[(data.Team == "India") & (data.Medal != "No Medal")]
```

```python
# Plotting the graphs.

sb.set(style = "darkgrid")
plt.figure(figsize = (20, 10))
sb.countplot(x = 'Year', data = womenInOlympics)
plt.title("Participation of Women in Olympics")
```

```python
# Plotting the graphs.

sb.set(style = "darkgrid")
plt.figure(figsize = (20, 10))
sb.countplot(x = 'Year', data = menInOlympics)
plt.title("Participation of Men in Olympics")
```

```python
sb.set(style = "darkgrid")
plt.figure(figsize = (20, 10))
sb.countplot(x = "Year", data = womenGoldAboveTwentyfive)
plt.title("Women Gold Medallists Above the Age of TwentyFive")
```

```python
sb.set(style = "darkgrid")
plt.figure(figsize = (20, 10))
sb.countplot(x = "Year", data = menGoldAboveThirty)
plt.title("Men Gold Medallists Above the Age of Thirty")
```

```python
sb.set(style = "darkgrid")
plt.figure(figsize = (20, 10))
sb.countplot(x = "Year", data = medalsIndia)
plt.title("Medals Secured By India in Olympics")
```

```python
# Standardizing the data 

# Printing mean values
print("Average height of participants in the Olympics is : ", data.Height.mean())
print("Average weight of participants in the Olympics is : ", data.Weight.mean())
print("Average age of participants in the Olympics is : ", data.Age.mean())

# Printing standard deviation
print("\nStandard Deviation in height of participants in the Olympics is : ", data.Height.std())
print("Standard Deviation in weight of participants in the Olympics is : ", data.Weight.std)
print("Standard Deviation in age of participants in the Olympics is : ", data.Age.std())

# Printing variance 
print("\nVariance in height of participants in the Olympics is : ", data.Height.mean())
print("Variance in weight of participants in the Olympics is : ", data.Weight.var())
print("Variance in age of participants in the Olympics is : ", data.Age.var())
```

```python
# Normalizing the data by printing the values between 0 and 1

print("Normalized data for height : ")
print((data.Height - data.Height.min())/(data.Height.max() - data.Height.min()))
print("\n")

print("Normalized data for weight : ")
print((data.Weight - data.Weight.min())/(data.Weight.max() - data.Weight.min()))
print("\n")

print("Normalized data for age : ")
print((data.Age - data.Age.min())/(data.Age.max() - data.Age.min()))
```
